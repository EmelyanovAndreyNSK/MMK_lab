Равномерное нормальное давление, приложенное к обводу кругового отверстия (стр. 195, 3)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class TaskParameters:
  def __init__(self, Pressure, Radius):
    self.Pressure = Pressure    
    self.Radius = Radius  

def Qxx(x, y, parameters):
    return -parameters.Pressure * parameters.Radius ** 2 * (x ** 2 - y ** 2) / (x ** 2 + y ** 2) ** 2

def Qyy(x, y, parameters):
    return -Qxx(x, y, parameters)

def Qxy(x, y, parameters):
    return -2 * parameters.Pressure * parameters.Radius ** 2 * x * y / (x ** 2 + y ** 2) ** 2     

parameters = TaskParameters(5., 1.)

xAxis = np.arange(-3, 3, 0.01)
yAxis = np.arange(-3, 3, 0.01) 

qxx = np.zeros((len(xAxis), len(yAxis)))
qyy = np.zeros((len(xAxis), len(yAxis)))
qxy = np.zeros((len(xAxis), len(yAxis)))

for i in range(1, len(xAxis)):
  for j in range(1, len(yAxis)):
    x = xAxis[i]
    y = yAxis[j]  
    if x ** 2 + y ** 2 < parameters.Radius:
        continue          
    qxx[i, j] = Qxx(x, y, parameters)
    qyy[i, j] = Qyy(x, y, parameters)
    qxy[i, j] = Qxy(x, y, parameters) 

def heatmap2d(arr: np.ndarray, title):
    plt.imshow(arr, cmap='viridis')
    plt.colorbar()
    plt.title(title)
    plt.show()
```

```python
heatmap2d(qxx, "Qxx")
heatmap2d(qyy, "Qyy")
heatmap2d(qxy, "Qxy")
```
